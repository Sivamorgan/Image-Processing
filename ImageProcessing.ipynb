## 2. Image Tensor Manipulation with numpy 

In this task, you will work with image data represented as NumPy arrays. You will perform basic image manipulation operations using NumPy, such as flipping, rotating, and resizing images. **Objective:**

**Requirements:**

2. **Rotate the Image**: Rotate the image 90 degrees clockwise
3. **Convert to Grayscale**: Convert the color image to grayscale.
4. **Increase Image Brightness**: Increase the brightness of the image by 10%.
5. **Decrease Image Brightness**: Decrease the brightness of the image by 10%.
6. **Increase Image Contrast**: Increase the contrast of the image by 10%.
7. **Decrease Image Contrast**: Decrease the contrast of the image by 10%.

**Expected Output:**
- The script should sequentially perform the mentioned operations and display the manipulated image at the end of each task.


````python
# Open an image file and convert to numpy array
import numpy as np
from PIL import Image
imag=PIL.Image.open(".\Pill.jpg")
numpy_array_image=np.array(imag)

# Rotate the image 90 degrees clockwise
rotated_array=np.rot90(numpy_array_image,3)
rotated_image=Image.fromarray(rotated_array)
rotated_image

# Convert to grayscale
gray_array=np.dot(numpy_array_image[...,:3],[0.2989, 0.5870, 0.1140])
gray_image=Image.fromarray(gray_array)

# Increase the brightness by 10%
brightened_image_array=np.round(numpy_array_image+0.1).astype(np.uint8)
brightened_image=Image.fromarray(brightened_image_array)

# Decrease the brightness by 10%
brightened_image_array=np.round(numpy_array_image-0.1).astype(np.uint8)
brightened_image_minus=Image.fromarray(brightened_image_array)

# Increase the contrast by 10%
minval = np.percentile(numpy_array_image, 2)
maxval = np.percentile(numpy_array_image, 98)
contrast_range = maxval - minval
increase_decrease_amount = 0.1 * contrast_range
new_minval = minval - increase_decrease_amount / 2
new_maxval = maxval + increase_decrease_amount / 2
numpy_array_image_1 = np.clip(numpy_array_image, new_minval, new_maxval)
contrast_increased_array = ((numpy_array_image_1 - new_minval) / (new_maxval - new_minval)) * 255
Image.fromarray(contrast_increased_array.astype(np.uint8))

# Decrease the contrast by 10%
new_minval = minval + increase_decrease_amount / 2
new_maxval = maxval - increase_decrease_amount / 2
numpy_array_image_2 = np.clip(numpy_array_image, new_minval, new_maxval)
contrast_decreased_array = ((numpy_array_image_2 - new_minval) / (new_maxval - new_minval)) * 255
Image.fromarray(contrast_decreased_array.astype(np.uint8))
````
